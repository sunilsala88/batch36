```python
df['Date added[16]']=df['Date added[16]'].str.replace(' ','-')
df
```

```python
l1=[]
for i in df['Date added[16]']:
    if i[-3]=='[':
        l1.append(i[:-3])
    else:
        l1.append(i)
l1
df['Date added[16]']=l1
```

```python
from io import StringIO
import ssl
import urllib.request
import certifi

url = 'https://www.screener.in/screens/281185/nifty-50/'
ssl_context = ssl.create_default_context(cafile=certifi.where())

with urllib.request.urlopen(url, context=ssl_context) as response:
    html = response.read().decode('utf-8')

df9 = pd.read_html(StringIO(html))
df=df9[0]
df
```

```python
df.info()
```

```python
df=df.drop(15,axis=0)
df
```

```python
df['P/E']=df['P/E'].astype('float')
df.info()
```
